## Divide between Features and Target

```python
df_columns = df.columns

predictors = df[df_columns[df_columns != 'Strength']] # all columns except Strength
target = df['Strength'] # Strength column
```

```python
n_cols = predictors.shape[1]
```

## **Keras Model Task A**

```python
import keras
from keras.layers import Dense
from keras.models import Sequential
```

```python
def regression_model():
  model = Sequential()
  model.add(Dense(10,activation = 'relu', input_shape = (n_cols,)))
  model.add(Dense(1))
  #
  model.compile(optimizer = 'adam', loss = 'mean_squared_error')
  return model
```

## Keras Model Task A Number 1
Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the train_test_splithelper function from Scikit-learn.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size = 0.3, random_state = 1)
```

## Keras Model Task A Number 2
Train the model on the training data using 50 epochs.



```python
model = regression_model()
model.fit(X_train,y_train,epochs = 50)
```

## Keras Model Task A Number 3
Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

```python
from sklearn.metrics import mean_squared_error
y_hat = model.predict(X_test)
print(mean_squared_error(y_test, y_hat))
```

The Mean Square Error is 152.45


## Keras Model Task A Number 4
Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

```python
MSE_List = []
for x in range(50):
  model = regression_model()
  model.fit(X_train,y_train,epochs = 50)
  y_hat = model.predict(X_test)
  loss = mean_squared_error(y_test, y_hat)
  MSE_List.append((x,loss))

print(MSE_List)


```

## Keras Model Task A Number 5
Report the mean and the standard deviation of the mean squared errors.

```python
import numpy as np
print("The Mean of MSE is {}".format(np.mean(MSE_List)))
print("The Std of MSE is {}".format(np.std(MSE_List)))
```

## **Keras Model Task B**
In this model will normalize the data using help of StandardScaler from sklearn.
Recall that one way to normalize the data is by subtracting the mean from the individual predictors and dividing by the standard deviation and its the same as StandardScaler from sklearn

```python
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
SS = StandardScaler()
X_train_norm = SS.fit_transform(X_train)
X_test_norm = SS.transform(X_test)
```

```python
MSE_List = []
for x in range(50):
  model = regression_model()
  model.fit(X_train_norm,y_train,epochs = 50)
  y_hat = model.predict(X_test_norm)
  loss = mean_squared_error(y_test, y_hat)
  MSE_List.append((loss))
```

```python
print("The Mean of MSE is {}".format(np.mean(MSE_List)))
print("The Std of MSE is {}".format(np.std(MSE_List)))
```

## **Keras Model Task D**
Repeat Part B but use a neural network with the following instead:

- Three hidden layers, each of 10 nodes and ReLU activation function.

```python
SS = StandardScaler()
X_train_norm = SS.fit_transform(X_train)
X_test_norm = SS.transform(X_test)
```

```python
def regression_model():
  model = Sequential()
  model.add(Dense(10,activation = 'relu', input_shape = (n_cols,)))
  model.add(Dense(10,activation = 'relu', input_shape = (n_cols,)))
  model.add(Dense(10,activation = 'relu', input_shape = (n_cols,)))
  model.add(Dense(1))
  #
  model.compile(optimizer = 'adam', loss = 'mean_squared_error')
  return model

MSE_List = []
for x in range(50):
  model = regression_model()
  model.fit(X_train_norm,y_train,epochs = 50)
  y_hat = model.predict(X_test_norm)
  loss = mean_squared_error(y_test, y_hat)
  MSE_List.append((x,loss))
```

```python
print("The Mean of MSE is {}".format(np.mean(MSE_List)))
print("The Std of MSE is {}".format(np.std(MSE_List)))
```
